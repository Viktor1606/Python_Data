```python
t = np.linspace(0,10,51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.axis([0.5,9.5, -2.5,2.5])

plt.plot(t,f, color='green')
plt.title('График f(t)',fontsize=20, fontweight='bold', color='blue')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
y1
```

```python
y2 = 2 * x + 0.5
y2
```

```python
y3 = -3 * x -1.5
y3
```

```python
y4 = np.sin(x)
y4
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x,y1)
ax2.plot(x,y2)
ax3.plot(x,y3)
ax4.plot(x,y4)
fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
ax1.set_xlim([-5,5])
ax1.set_title('График y1')
ax2.set_title('График y2')
ax3.set_title('График y3')
ax4.set_title('График y4')
plt.show()
```

```python
card = pd.read_csv('creditcard.csv')
```

```python
class_list = card['Class'].value_counts()
class_list
```

```python
class_list.plot(kind='barh')
plt.show()
```

```python
class_list.plot(kind='barh', logx = True)
plt.show()
```

```python
class0 = card.loc[card['Class'] == 0, ['V1']]
class1 = card.loc[card['Class'] == 1, ['V1']]
```

```python
plt.hist(class0['V1'], bins=20, density = True, alpha=0.5, label = 'Class 0', color = 'grey')
plt.hist(class1['V1'], bins=20, density = True, alpha=0.5, label = 'Class 1', color = 'red')
plt.legend()
plt.show()
```
